```python
# Import datasets
hki_data = pd.read_csv('44-helsinki.csv')
espoo_data = pd.read_csv('44-espoo.csv')
```

```python
# Merge datasets
full_dataset = pd.merge(hki_data, espoo_data, on = 'date', how = 'outer')
```

```python
full_dataset
```

```python
full_dataset_no_na = full_dataset.fillna(0)
```

```python
print(f"Days of observations: {len(full_dataset_no_na)}")
```

```python
# Get observations per street
obvs_per_street = full_dataset_no_na.drop(columns = ['date'])
obvs_per_street.sum().astype(int).sort_values()
```

```python
# Most cyclists
print(obvs_per_street.sum().sort_values().tail(1))
```

```python
simultaneous_observations = full_dataset.dropna()
print(f"There were simultaneous observations on {simultaneous_observations['date'].count()} days")
```

```python
obvs_per_street_no_na = full_dataset.drop(columns = ['date']).dropna()
obvs_per_street_no_na.sum().astype(int).sort_values()
```

There were some changes to the busiest streets, as expected.
