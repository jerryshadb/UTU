1. Apply a multivariable test and obtain P-value for

a) 2,3,5,4,4,3 4,2,3,5,2,3 3,1,4,4,3,5

b) Fi,Sw,Fi,No,Sw,Fi No,Sw,No,Fi,Fi,Fi Sw,Fi,No,Sw,Sw,No


```python
dataset_a = np.array([2,3,5,4,4,3,4,2,3,5,2,3,3,1,4,4,3,5])
dataset_b = np.array(["Fi","Sw","Fi","No","Sw","Fi","No","Sw","No","Fi","Fi","Fi", "Sw","Fi","No","Sw","Sw","No"])
```

```python
# To perform a multivariable test for data a) we need to split it into >2 sets, so 3 equal sized sets
dataset_a_1, dataset_a_2, dataset_a_3 = np.split(dataset_a, len(dataset_a) // 6)
```

```python
# verify partition
print(dataset_a_1)
print(dataset_a_2)
print(dataset_a_3)
```

```python
# check for normality
print(f"P-value from Shapiro Wilk test is {stats.shapiro(dataset_a_1).pvalue:.2f}")
print(f"P-value from Shapiro Wilk test is {stats.shapiro(dataset_a_2).pvalue:.2f}")
print(f"P-value from Shapiro Wilk test is {stats.shapiro(dataset_a_3).pvalue:.2f}")
```

```python
# ANOVA test
pvalue = stats.f_oneway(dataset_a_1, dataset_a_2, dataset_a_3)
print(f"P-value for dataset a) with ANOVA is {pvalue.pvalue:.2f}")
```

3. In this exercise we consider file simulated_data_2_5.csv. It has four groups of 100

observation: F, G, H and I. Apply multivariable test and obtain P-value

a) to compare columns F, G and H.

b) to compare columns F, G and I.


```python
data = pd.read_csv('simulated_data_2_5.csv')
data
```

```python
pvalue = stats.f_oneway(data['F'], data['G'], data['H'])
print(f"P-value for dataset 3 a) with ANOVA is {pvalue.pvalue:.2f}")
```

```python
pvalue = stats.kruskal(data['F'], data['G'], data['I'])
print(f"P-value for dataset 3 b) with Kruskaali is {pvalue.pvalue:.20f}")
```
