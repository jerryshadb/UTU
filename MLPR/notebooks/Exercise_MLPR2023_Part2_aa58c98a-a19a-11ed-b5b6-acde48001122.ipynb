Yes. There seems to be quite clear clustering apart from a few outliers. As stated in the research paper's results, there were a few mismatches of Arborio for Jasmine and Jasmine for Basmati. These can be clearly seen in the cluster as well and are almost expected to be mislabeled. 

```python
for n_components in range(1, 20):
    pca = PCA(n_components = n_components)
    components = pca.fit_transform(data[features])
    percent = np.sum(pca.explained_variance_ratio_)
    if percent >= 0.99:
        print(f"To cover 99 % of the variance, {n_components} components is needed")
        break
```

###### KNN

```python
'''
INPUT: 
X = training data

Y = targets

model = knn

OUTPUT:
Results of RepeatedKFold CV for model on data

'''

def get_knn_performance(X, y, model):
    predictions = list()
    actuals = list()
    rkf = RepeatedKFold(n_splits = 5, n_repeats = 3)
    
    for train_idx, test_idx in rkf.split(X):
        x_train, x_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        model.fit(x_train, y_train)
    return model.score(x_test, y_test)
```

```python
# get best K for KNN
performances = dict()
for neighbours in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors = neighbours)
    performances[neighbours] = get_knn_performance(data[features].values, data['specie'].values, knn)
```

```python
performances
```

```python
print(f"Best K is K={(max(performances, key=performances.get))}")
```

```python
best_hyperparam = (max(performances, key=performances.get))
value = performances[(max(performances, key=performances.get))]

plt.plot(performances.keys(), performances.values())
plt.xlabel('Hyperparameter')
plt.ylabel('Accuracy')
plt.title('Accuracy versus Hyperparameter')
plt.scatter(best_hyperparam, value, color='red')

plt.show()
```

After a few runs, it'd seem KNN is very prone to overfitting as a chance for 1.0 accuracy is quite high. Plot reference: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html

###### Random Forest

```python
def get_forest_params(X, y):
    rkf = RepeatedKFold(n_splits = 5, n_repeats = 3)
    rfc = RandomForestClassifier()
    parameters = {
        'max_depth': [5, 10, 15, 20],
        'max_features': ['auto', 'sqrt']
    }
    grid_search = GridSearchCV(rfc, parameters, scoring = 'accuracy', cv = rkf)
    grid_search.fit(X, y)
    
    return grid_search
```

```python
forest_params = get_forest_params(data[features].values, data['specie'].values)
print(f"Best params and score for Random Forest:\nParameters: {forest_params.best_params_}\nScore: {forest_params.best_score_:.2f}")
results_forest_df = pd.DataFrame(data = forest_params.cv_results_)
```

#### Feature importance

```python
best_random_forest = forest_params.best_estimator_
feature_importances = best_random_forest.feature_importances_
pd.DataFrame({'feature': features, 'importance': feature_importances})
```

###### Multi-Layer Perceptron

```python
# referenced from: https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html
def get_mlp_params(X, y):
    rkf = RepeatedKFold(n_splits = 5, n_repeats = 3)
    mlp = MLPClassifier(early_stopping = True) 
    
    parameters = {
        'hidden_layer_sizes': [(25, ), (50, ), (75, ), (80, ), (95, ), (120,), (150, ), (200, ), (210, )],
        'activation': ['logistic', 'relu'],
        'solver': ['adam', 'sgd'],
        'validation_fraction': [0.1, 0.5]
    }
    
    grid_search = GridSearchCV(mlp, parameters, scoring = 'accuracy', cv = rkf)
    grid_search.fit(X, y)
    
    return grid_search
```

```python
mlp_params = get_mlp_params(data[features].values, data['specie'].values)
print(f"Best params and score for Random Forest:\nParameters: {mlp_params.best_params_}\nScore: {mlp_params.best_score_:.2f}")
```
